```python
import pandas as pd
import os
```

```python
def get_paperid(filename):
    return filename.split('.')[0].split('_')[1]
```

```python
def get_candidate(folder_path):
    return [f for f in os.listdir(folder_path) if f.endswith('.candidate')]
```

```python
def evaluate(folder_path, df):
    candidates = get_candidate(folder_path)
    print(len(candidates))
    for candidate in candidates:
        paperid = get_paperid(candidate)
        f = open("{}/{}".format(folder_path, candidate), "r")
        summary = f.read()
        df.loc[df['paper_id']==int(paperid), 'summary'] = summary
```

```python
def print_result(df, i):
    print('-------{} summary---------'.format(i))
    print(df['summary'].iloc[i])
    print('------conclusion--------')
    print(df['raw_conclusion'].iloc[i])
```

```python
quarantine_df = pd.read_csv('quarantine_final.csv')
quarantine_df
```

```python
evaluate('./quarantine_final', quarantine_df)
```
